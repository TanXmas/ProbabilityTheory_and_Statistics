## <font color = DarkGreen size = 5> <b>Непрерывная случайная величина</b> </font>

Непрерывная случайная величина может принять любое действительное значение из некоторого промежутка ненулевой длины, что делает невозможным её представление в виде таблицы (т.к. действительных чисел несчётно много). В этой связи непрерывную случайную величину задают функциями двух типов: **функцией распределения** и **функцией плотности распределения**.

```python
import numpy as np
#import pandas as pd

from scipy import stats

import matplotlib.pyplot as plt
#import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

***
### <font color = MediumBlue size = 4> <b>Функция распределения (cdf - cumulative distribution function)</b> </font>

<u>**Функция распределения**</u> непрерывной случайной величины $X$ определяется как \
$F(X)=P(X \leq x) = \sum{P(X=k) \cdot [X \leq x]}$, где \
$[X \leq x] = 
\begin{cases}
1 & \text{if } X \leq x \\
0 & \text{if } X > x
\end{cases}
$

**Свойтсва функции распределения**:
1) Значения функции распределения принадлежат отрезку $[0,1]$: $0 \leq F(x) \leq 1$

2) $F(x)$ – неубывающая функция: $F(x_2) \geq F(x_1)$, если $x_2 > x_1$
    
* Вероятность того, что случайная величина примет значение, заключенное в интервале $(а, b)$, равна приращению функции распределения на этом интервале: $P(a \leq X < b) = F(b) - F(a)$
* Вероятность того, что непрерывная случайная величина X примет одно определенное значение, равна нулю.

3) Если возможные значения случайной величины принадлежат интервалу $(а, b)$, то: \
&emsp;1) $F(x)=0$, при $x < a$ \
&emsp;2) $F(x)=1$, при $x \geq a$

* Если возможные значения непрерывной случайной величины расположены на всей оси $x$, то справедливы следующие предельные соотношения: \
&emsp;1) $\displaystyle \lim_{x\rightarrow -\infty}{F(x)=0}$ \
&emsp;2) $\displaystyle \lim_{x\rightarrow \infty}{F(x)=1}$

___
### <font color = MediumBlue size = 4> <b>Функция плотности распределения / дифференциальная функция (pdf - probability density function)</b> </font>

Плотностью распределения вероятностей непрерывной случайной величины $X$ называют функцию \
$f(x) = F'(X)$

>Из этого определения следует, что функция распределения является первообразной для плотности распределения.

**Свойства плотности распределения**:
1) Плотность распределения – неотрицательная функция: $f(x) \geq 0$

2) Несобственный интеграл от плотности распределения в пределах от $-\infty$ до $\infty$ равен единице: $\displaystyle \int_{-\infty}^{\infty}{f(x)dx=1}$
> Геометрически это означает, что вся площадь криволинейной трапеции, ограниченной осью Ох и кривой распределения, равна единице. В частности, если все возможные значения случайной величины принадлежат интервалу $(a,b)$, то $\displaystyle \int_{a}^{b}{f(x)dx=1}$


#### <font color = SteelBlue size = 3> <b>Вероятность попадания непрерывной случайной величины в заданный интервал</b> </font>

Вероятность того, чти непрерывная случайная величина $X$ примет значение, принадлежащее интервалу $(a,b)$, равна определенному интегралу от плотности распределения, взятому в пределах от $a$ до $b$: \
$\displaystyle P(a<X<b) = \int_{a}^{b}{f(x)dx}$

Геометрически полученный результат можно истолковать так: вероятность того, что непрерывная случайная величина примет значение, принадлежащее интервалу $(a,b)$, равна площади криволинейной трапеции, ограниченной осью $Ox$, кривой рапределения $f(x)$ и прямыми $x=a$ и $x=b$.


#### <font color = SteelBlue size = 3> <b>Вероятностный смысл плотности распределения</b> </font>

Пусть $F(x)$ – функция распределения непрерывной случайной величины $X$. По определению плотности распределения $f(x) = F'(X)$ или в иной форме:\
$\displaystyle f(x) = \lim_{\Delta{x} \rightarrow 0}{\frac{F(x+\Delta{x})-F(x)}{\Delta{x}}}$

Как уже известно, разность $F(x+\Delta{x})-F(x)$ определяет вероятность того, что $X$ примет значение, принадлежащее интервалу $(x,\Delta{x})$. Таким образом, предел отношения вероятности того, что непрерывная случайная величина примет значение, принадлежащее интервалу $(x,\Delta{x})$, к длине этого интервала (при $\Delta{x} \rightarrow 0$) равен значению плотности распределения в точке $x$.\
**Итак, функция $f(x)$ определяет плотность распределения вероятности для каждой точки $x$.**

___
### <font color = MediumBlue size = 4> <b>Математическое ожидание непрерывной случайной величины</b> </font>

Пусть непрерывная случайная величина $X$ задана плотностью распределения $f(x)$. Допустим, что все возможные значения $X$ принадлежат отрезку $[a,b]$. Разобьем этот отрезок на $n$ частичных отрезков длиной $\Delta{x_1},\Delta{x_2},\ldots,\Delta{x_n}$ и выберем в каждом из них произвольную точку $x_i (i = 1,2,\ldots,n)$. \
Cоставим сумму произведений возможных значений $x_i$ на вероятности попадания их в интервал $\Delta{x_i}$ (произведение $f(x)\Delta{x}$ приближенно равно вероятности попадания $X$ в интервал $\Delta{x}$): \
$\displaystyle \sum{x_i f(x_i) \Delta{x_i}} = \int_{a}^{b}{x f(x) dx}$

Математическим ожиданием непрерывной случайной величины $X$, возможные значения которой принадлежат отрезку $[a,b]$, называют определенный интеграл: $\displaystyle M(X) = \int_{a}^{b}{x f(x) dx}$

Если возможные значения $X$ принадлежат всей оси $Ox$, то $\displaystyle M(X) = \int_{-\infty}^{\infty}{x f(x) dx}$

___
### <font color = MediumBlue size = 4> <b>Дисперсия непрерывной случайной величины</b> </font>

Дисперсией непрерывной случайной величины называют математическое ожидание квадрата ее отклонения. \

Если возможные значения $X$ принадлежат отрезку $[a,b]$, то $\displaystyle D(X) = \int_{a}^{b}{[x-M(X)]^2 f(x) dx}$

Если возможные значения $X$ принадлежат всей оси $Ox$, то $\displaystyle D(X) = \int_{-\infty}^{\infty}{[x-M(X)]^2 f(x) dx}$

### <font color = MediumBlue size = 4> <b>Среднее квадратическое отклонение непрерывной случайной величины</b> </font>

$\sigma(X) = \sqrt{D(X)}$

### <font color = MediumBlue size = 4> <b>Закон равномерного распределения вероятностей</b> </font>

Распределение вероятностей называют <u>**равномерным**</u>, если на интервале, которому принадлежат все возможные значения случайной величины, плотность распределения сохраняет постоянное значение.

Рассмотрим некоторый конечный промежуток, пусть для определённости это будет отрезок $[a;b]$. Если случайная величина $X$ обладает постоянной плотностью распределения вероятностей на данном отрезке и нулевой плотностью вне него, то говорят, что она распределена равномерно. \
При этом функция плотности будет строго определённой:
$f(x) =
\begin{cases}
0 & \quad \text{if } x \leq a \\
\displaystyle{\frac{1}{b-a}} & \quad \text{if }  a < x \leq b \\
0 & \quad \text{if } x > b
\end{cases}$

#### <font color = SteelBlue size = 3> <b>Математическон ожидание</b> </font>

$\displaystyle M(X) = \frac{a+b}{2}$

#### <font color = SteelBlue size = 3> <b>Дисперсия</b> </font>

$\displaystyle D(X) = \frac{(b-a)^2}{12}$


```python
rv = stats.uniform()

plt.figure(figsize = [6*2, 4])

x = np.linspace(-1, 2)
# функция распределения
cdf = rv.cdf(x)
plt.subplot(1, 2, 1)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('функция распределения')
plt.grid()

# функция плотности распределения
pdf = rv.pdf(x)
plt.subplot(1, 2, 2)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('функция плотности распределения')
plt.grid();
```

### <font color = MediumBlue size = 4> <b>Нормальный закон распределения вероятностей</b> </font>

<u>**Нормальным**</u> называют распределение вероятностей непрерывной случайной величины, которое описывается **плотностью** \
<font size = 4>
$\displaystyle f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-a)^2}{2\sigma^2}}$
</font>, где $a = M(X) \;\;\; \sigma^2 = D(X)$

<u>**Общим**</u> называют **нормальное распределение** с произвольными параметрами $a$ и $\sigma$ $(\sigma > 0)$.

<u>**Нормированным**</u> называют **нормальное распределение** с параметрами $a=0$ и $\sigma=1$.

#### <font color = SteelBlue size = 3> <b>Нормальная кривая</b> </font>

График плотности нормального распределения называют нормальной кривой (кривой Гаусса). \
<font size = 4>
$\displaystyle y = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-a)^2}{2\sigma^2}}$
</font>
<br>
1) Очевидно, функция определена на всей оси $x$.

2) При всех значениях $x$ функция принимает положительные значения, т.е. нормальная кривая расположена над осью $Ox$.

3) Предел функции при неограниченном возрастании $x$ (по абсолютной величине) равен нулю $\lim\limits_{|x| \to \infty} y=0$, т.е. ось $Ox$ служит горизонтальной асимптотой графика.

4) Исследуем функциф на экстремум. Найдем первую производную: \
<font size = 4> $\displaystyle y\,' = -\frac{x-a}{\sigma^3\sqrt{2\pi}} e^{-\frac{(x-a)^2}{2\sigma^2}}$ </font> \
<br>
$\begin{cases}
y\,' = 0 & \quad \text{if } x = a \\
y\,' > 0 & \quad \text{if } x < a \\
y\,' < 0 & \quad \text{if } x > a
\end{cases}$ \
Следовательно, при $x=a$ функция имеет максимум, равный $\displaystyle \frac{1}{\sigma\sqrt{2\pi}}$

5) Разность $x-a$ содержится в аналитическом выражении функции в квадрате, т. е. график функции симметричен относительно прямой $x-a$.

6) Исследуем функцию на точки перегиба. Найдем вторую производную:
<font size = 4> $\displaystyle y\,'' = -\frac{1}{\sigma^3\sqrt{2\pi}} e^{-\frac{(x-a)^2}{2\sigma^2}} \left[1 - \frac{(x-a)^2}{\sigma^2}\right]$ </font>
<br>
$\displaystyle y\,'' = 0 \text{ if } x = a \pm \sigma$ и при переходе через эти точки она меняет знак. В обеих этих точках значение функции равно $\displaystyle \frac{1}{\sigma\sqrt{2\pi e}}$. Таким образом, точки графика $\displaystyle \left(a-\sigma,\frac{1}{\sigma\sqrt{2\pi e}}\right)$ и $\displaystyle \left(a+\sigma,\frac{1}{\sigma\sqrt{2\pi e}}\right)$ являются точками перегиба.
<br><br>

```python
plt.figure(figsize = [6*2, 4])

Mx, sigma =0, 1
rv = stats.norm(loc=Mx, scale=sigma)
x = np.linspace(-3, 3, 100)

# функция распределения
plt.subplot(1, 2, 1)
plt.plot(x, rv.cdf(x))
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('функция распределения')
plt.grid()

# функция плотности распределения
plt.subplot(1, 2, 2)
plt.plot(x, rv.pdf(x))
plt.scatter([Mx], [1 / (sigma * np.sqrt(2 * np.pi))], color="red"); # максимум
plt.scatter([Mx-S], [1 / (sigma * np.sqrt(2 * np.pi * np.e))], color="red"); # точка перегиба
plt.scatter([Mx+S],  [1 / (sigma * np.sqrt(2 * np.pi * np.e))], color="red"); # точка перегиба

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('функция плотности распределения')
plt.grid();
```

#### <font color = SteelBlue size = 3> <b>Влияние параметров нормального распределения на форму нормальной кривой</b> </font>

Изменение величины параметра $a$ (математического ожидания) не изменяет формы нормальной кривой, а приводит лишь к ее сдвигу
вдоль оси $Ox$: вправо, если $a$ возрастает, и влево, если $a$ убывает.
<br><br>

Максимум дифференциальной функции нормального распределения равен $\displaystyle \frac{1}{\sigma\sqrt{2\pi}}$.\
Следовательно, с возрастанием $\sigma$ максимальная ордината нормальной кривой убывает, а сама кривая становится более пологой, т.е. сжимается к оси $Ox$; при убывании $\sigma$ нормальная кривая становится более «островершинной» и растягивается в положительном направлении оси $Oy$.
<br><br>

При любых значениях параметров $a$ и $\sigma$ площадь, ограниченная нормальной кривой и осью $x$, остается равной единице.
<br><br>

#### <font color = SteelBlue size = 3> <b>Вычисление вероятности заданного отклонения</b> </font>

Часто требуется вычислить вероятность того, что отклонение нормально распределенной случайной величины $X$ по абсолютной величине меньше заданного положительного числа $\delta$, т. е. требуется найти вероятность осуществления неравенства $|X-a|<\delta$.

$\displaystyle P(|X-a|<\delta) = P(-\delta<X-a<\delta) = P(a-\delta<X<a+\delta) = 
\Phi\left(\frac{(a+\delta)-a}{\sigma}\right) - \Phi\left(\frac{(a-\delta)-a}{\sigma}\right) =
\Phi\left(\frac{\delta}{\sigma}\right) - \Phi\left(-\frac{\delta}{\sigma}\right)$

Учитывая, что функция Лапласа — нечетная: 
$\displaystyle \Phi\left(-\frac{\delta}{\sigma}\right) = - \Phi\left(\frac{\delta}{\sigma}\right)$

$$\displaystyle P(|X-a|<\delta) = 2\Phi\left(\frac{\delta}{\sigma}\right)$$

#### <font color = SteelBlue size = 3> <b>Правило трех сигм</b> </font>

Практически достоверным является тот факт, что нормально распределённая случайная величина $X$ примет значение из промежутка $(a-3\sigma,a+3\sigma)$.

$\displaystyle P(|X-a|<\delta) = 2\Phi\left(\frac{\delta}{\sigma}\right)$

Для $\delta = \sigma t$: $P(|X-a|<\sigma t) = 2\Phi(t)$

Тогда, вероятность отклонения от матожидания менее чем на $t=3$ и следовательно $\sigma t = 3\sigma$ составляет: \
$\displaystyle P(|X-a|<3\sigma)=2\Phi\left(\frac{3\sigma}{\sigma}\right)=2\Phi(3)=0.9973$

#### <font color = SteelBlue size = 3> <b>Центральная предельная теорема</b> </font>

Если случайная величина $X$ представляет собой сумму очень большого числа взаимно независимых случайных величин, влияние каждой из которых на всю сумму ничтожно мало, то $X$ имеет распределение, близкое к нормальному.

### <font color = MediumBlue size = 4> <b>Показательное / экспоненциальное распределение вероятностей</b> </font>

**Показательным** или **экспоненциальным** называют распределение, которое характеризуется следующей функцией плотности: \
$\begin{cases}
0 & \quad \text{if } x < 0 \\
\lambda e^{-\lambda x}  & \quad \text{if } x \leq a
\end{cases}$, где $\lambda > 0$

$M(X) = \cfrac{1}{\lambda}$

$D(X) = \cfrac{1}{\lambda^2}$

```python
plt.figure(figsize = [6*2, 4])

params = [1, 2, 5, 10]
x = np.linspace(0, 20, 100)

for p in params:
    rv = stats.expon(scale=p)
    label = f'$\lambda = {p}$'
    plt.subplot(1, 2, 1)
    plt.plot(x, rv.cdf(x), label=label)
    
    plt.subplot(1, 2, 2)
    plt.plot(x, rv.pdf(x), label=label)

plt.subplot(1, 2, 1)
plt.legend()
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('функция распределения')

plt.subplot(1, 2, 2)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('функция плотности распределения');
```
