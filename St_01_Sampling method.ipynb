# <font color = Crimson size = 6> <b>Выборочный метод</b> </font>

***
## <font color = DarkGreen size = 5> <b>Генеральная и выборочная совокупности</b> </font>

<u>**Генеральная совокупность**</u> — это множество, которое содержит данные обо всех объектах, соответствующих определенным характеристикам.

<u>**Выборка / выборочная совокупность**</u> — это случайным образом выбранная часть генеральной совокупности.

Исследуемое множество называют <u>**статистической совокупностью**</u>, а количество его элементов – <u>**объемом**</u> совокупности.

***
### <font color = MediumBlue size = 4> <b>Повторная и бесповторная выборки</b> </font>

При составлении выборки можно поступать двумя способами: после того как объект отобран и над ним произведено наблюдение, он может быть возвращен либо не возвращен в генеральную совокупность. В соответствии со сказанным выборки подразделяют на **повторные** и **бесповторные**.

<u>**Повторной**</u> называют выборку, при которой отобранный объект (перед отбором следующего) возвращается в генеральную совокупность.

<u>**Бесповторной**</u> называют выборку, при которой отобранный объект в генеральную совокупность не возвращается.

На практике обычно пользуются бесповторным случайным отбором.

***
### <font color = MediumBlue size = 4> <b>Репрезентативная выборка</b> </font>

Для того чтобы по данным выборки можно было достаточно уверенно судить об интересующем признаке генеральной совокупности, необходимо, чтобы объекты выборки правильно представлять пропорции генеральной совокупности, т.е. выборка должна быть **репрезентативной (представительной)**.

<u>**Репрезентативная выборка**</u> — часть генеральной совокупности, являющаяся моделью генеральной совокупности и отражющая свойства генеральной совокупности.

В силу закона больших чисел можно утверждать, что выборка будет репрезентативной, если ее осуществить случайно: каждый объект выборки отобран случайно из генеральной совокупности, если все объекты имеют одинаковую вероятность попасть в выборку.

#### <font color = SteelBlue size = 3> <b>Способы формирования репрезентативной выборки</b> </font>
1) **простая случайная выборка** (simple random sample) \
&nbsp; Cлучайным образом выбираются элементы генеральной совокупности для участия в эксперименте или исследовании. \
&nbsp; При увеличение объёма случайной выборки, она постепенно начинает напоминать генеральную совокупность по своим характеристикам.

2) **стратифицированная выборка** (stratified sample) \
&nbsp; Cначала генеральная совокупность разбивается на несколько обособленных и различных по своей природе страт-групп. \
&nbsp; Затем используется метод случайной выборки для каждой группы.

3) **кластерная / групповая выборка** (cluster sample) \
&nbsp; Cначала генеральная совокупность разбивается на несколько кластеров. Все кластеры очень похожи между собой. \
&nbsp; Выбирается несколько кластеров для участия в эксперименте или исследовании. \
&nbsp; Затем используется метод случайной выборки из вабранных кластеров.

#### <font color = SteelBlue size = 3> <b>Статистическое распределение выборки</b> </font>

Пусть из генеральной совокупности извлечена выборка, причем $x_1$ наблюдалось $n_1$ раз, $x_2$ — $n_2$ раз, $x_k$ — $n_k$ раз и $\sum{n_i} = n$ — объем выборки.

Наблюдаемые значения $x_i$ называют <u>**вариантами**</u>, а последовательность вариант, записанных в возрастающем порядке, — <u>**вариационным рядом**</u>.

Числа наблюдений называют <u>**частотами**</u>, а их отношения к объему выборки $\displaystyle \frac{n_i}{n} = W_i$ — <u>**относительными частотами**</u>.

<u>**Статистическим распределением**</u> выборки называют перечень вариант и соответствующих им частот или относительных частот. \
Статистическое распределение можно задать также в виде последовательности интервалов и соответствующих им частот (в качестве частоты, соответствующей интервалу, принимают сумму частот, попавших в этот интервал).

## <font color = DarkGreen size = 5> <b>Виды и источники статистических данных</b> </font>

**Типы данных:**
1. Пространственные (cross-sectional data) — серия из $n$ независимых наблюдений $(p+1)$-мерной случайной величины $(x_1, \ldots, x_p; Y)$.
2. Временные (time-series data) — серия наблюдений одной и той же случайной величины в последовательные моменты времени.
3. Пространственно-временные

**Типы переменных:**
1. Количественные (представляют собой непосредственно измерение значение некоторого признака)
    * непрерывные — переменные, принимающие любое значение на некотором промежутке.
    * дискретные — переменные, принимающие определенное значение на некотором промежутке.
2. Категориальные:
    * порядковые / ранговые (единственная арифметическая операция которая возможно в случае ранговых переменных - это сравнение)
    * номинальные / качественные (используются для разделения испытуемых или наблюдений на группы)

**Шкала измерения данных:**
1. Для качественных данных:
    * номинальная
    * порядковая
2. Для коиличественных данных:
    * интервальная
    * относительная

## <font color = DarkGreen size = 5> <b>Дискретный вариационный ряд</b> </font>

<u>**Дискретный вариационный ряд**</u> \
– это упорядоченное по возрастанию (как правило) множество вариант $x_1,x_2,\ldots,x_k$ (значений величины $X$) и соответствующих им частот либо относительных частот.

Частоты выборочной совокупности обозначают через $n_1,n_2,\ldots,n_k$, частоты генеральной совокупности – через $N_1,N_2,\ldots,N_k$.

Относительные частоты рассчитываются по формуле: $\displaystyle w_i = \frac{n_i}{n}$, где $\displaystyle n = \sum_{i=1}^{k}{n_i}$ – объем выборки. При этом $\displaystyle \sum_{i=1}^{k}{w_i} = 1$

Аналогично для генеральной совокупности: $\displaystyle W_i = \frac{N_i}{N}$, где $\displaystyle N = \sum_{i=1}^{k}{N_i}$ – объем выборки и $\displaystyle \sum_{i=1}^{k}{W_i} = 1$

>Такие ряды появляются в результате учёта дискретной характеристики статистической совокупности, причём, варианты ряда не отличаются большим разнообразием.

### <font color = MediumBlue size = 4> <b>Полигон частот</b> </font>

<u>**Полигон частот**</u> – это ломаная, соединяющая соседние точки $(x_i,n_i)$. \
Для построения полигона частот на оси абсцисс откладывают варианты $x_i$, а на оси ординат — соответствующие им частоты $n_i$.

<u>**Полигон относительных частот**</u> – это ломаная, соединяющая соседние точки $(x_i,w_i)$. \
Для построения относительных полигона частот на оси абсцисс откладывают варианты $x_i$, а на оси ординат — соответствующие им частоты $w_i$.

### <font color = MediumBlue size = 4> <b>Эмпирическая функция распределения</b> </font>

Пусть известно статистическое распределение частот количественного признака $X$. \
Введем обозначения: \
$n_x$ — число наблюдений, при которых наблюдалось значение признака, меньшее $x$ \
$n$ — общее число наблюдений (объем выборки) \
Относительная частота события $X < x$ равна $\cfrac{n_x}{n}$. \
Если $x$ изменяется, то изменяется и относительная частота, т.е. относительная частота $\cfrac{n_x}{n}$ есть функция от $x$. \
Так как эта функция находится эмпирическим (опытным) путем, то ее называют эмпирической.

<u>**Эмпирической функцией распределения**</u> (функцией распределения выборки) называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоту события $X<x$: \
$F^*(x) = \cfrac{n_x}{n}$, где \
$n_x$ — число вариант, меньших $x$ \
$n$ — объем выборки

В отличие от эмпирической функции распределения выборки функцию распределения $F(x)$ генеральной совокупности называют  [**теоретической функцией распределения**](https://github.com/TanXmas/ProbabilityTheory_and_Statistics/blob/main/PT_03_Continuous%20random%20variable.ipynb). Различие между эмпирической и теоретической функциями состоит в том, что теоретическая функция $F(x)$ определяет вероятность события $X<x$, а эмпирическая функция $F^*(x)$ определяет относительную частоту этого же события.

**Свойства эмпирической функции распределения**:
1. значения эмпирической функции принадлежат отрезку $[0,1]$ \
2. $F^*(x)$ — неубывающая функция \
3. если $x_1$ — наименьшая варианта, то $F^*(x)=0$, при $x \leq x_1$; \
если $x_k$ — наибольшая варианта, то $F^*(x)=1$, при $x > x_k$

>Эмпирическая функция распределения выборки служит для оценки теоретической функции распределения генеральной совокупности.

## <font color = DarkGreen size = 5> <b>Интервальный вариационный ряд</b> </font>

<u>**Интервальный вариационный ряд**</u> \
Исследуемая величина принимает слишком много различных значений. Зачастую появляется в результате измерения непрерывной характеристики изучаемых объектов.

Для исследования интервального вариационного ряда:
1) определяется интервал, в пределах которого варьируются значения. \
2) данный интервал делится на частичные интервалы: \
формула Стерджеса: $k = 1 + 3.322\cdot\lg{n}$ – оптимальное количество интервалов, при этом результат округляют до ближайшего левого целого значения. \
Длины частичных интервалов могут быть различны, но в большинстве случаев использует равноинтервальную группировку: \
$\displaystyle h = \frac{x_{max}-x_{min}}{k}$ \
3) по каждому интервалу подсчитываются частоты – количество вариант, которые в него попали $n_i$.

### <font color = MediumBlue size = 4> <b>Гистограмма частот</b> </font>

<u>**Гистограмма частот**</u> – это фигура, состоящая из прямоугольников, ширина которых равна длинам частичных интервалов $h$, а высота – соответствующим плотностям частот $\displaystyle \frac{n_i}{h}$.

Для построения гистограммы частот на оси абсцисс откладывают частичные интервалы, а над ними проводят отрезки, параллельные оси абсцисс на расстоянии $\displaystyle \frac{n_i}{h}$. \
Площадь $i$-го частичного прямоугольника равна $n_i$ — сумме частот вариант $i$-гo интервала; следовательно, **площадь гистограммы частот равна сумме всех частот, т. е. объему выборки.**
<br><br>

<u>**Гистограммой относительных частот**</u> называют ступенчатую фигуру, состоящую из прямоугольников, основаниями которых служат частичные интервалы длиною $h$, а высоты равны отношению $\displaystyle \frac{w_i}{h}$ (плотность относительной частоты).

Для построения гистограммы относительных частот на оси абсцисс откладывают частичные интервалы, а над ними проводят отрезки, параллельные оси абсцисс на расстоянии $\displaystyle \frac{w_i}{h}$. \
Площадь $i$-го частичного прямоугольника равна $w_i$ — относительной частоте вариант, попавших в $i$-й интервал. \
Следовательно, **площадь гистограммы относительных частот равна сумме всех относительных частот, т. е. единице.**

<u>**Полигон относительных частот**</u> – это ломаная, соединяющая соседние точки $(x_i,w_i)$, где $x_i$ – середины интервалов.

### <font color = MediumBlue size = 4> <b>Эмпирическая функция распределения</b> </font>

<u>**Эмпирической функцией распределения**</u> (функцией распределения выборки) называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоту события $X<x$: \
$F^*(x) = \cfrac{n_x}{n}$, где \
$n_x$ — число вариант, меньших $x$ \
$n$ — объем выборки

**Свойства эмпирической функции распределения**:
1) значения эмпирической функции принадлежат отрезку $[0,1]$ \
2) $F^*(x)$ — неубывающая функция \
3) $F^*(x)$ — непрерывная функция
