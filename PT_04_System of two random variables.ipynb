# <font color = Crimson size = 6> <b>Система двух случайных величин</b> </font>

До сих пор рассматривались случайные величины, возможные значения которых определялись одним числом. Такие величины называют **одномерными**.

Кроме одномерных случайных величин изучают величины, возможные значения которых определяются двумя, тремя, $\ldots$, $n$ числами. Такие величины называются соответственно **двумерными, трехмерными, $\ldots$, $n$-мерными**.

Будем обозначать через $(X,Y)$ двумерную случайную величину. Каждую из величин $X$ и $Y$ называют **составляющей (компонентой)**; обе величины $X$ и $Y$, рассматриваемые одновременно, образуют систему двух случайных величин. Аналогично $n$-мерную величину можно рассматривать как систему $n$ случайных величин.

### <font color = MediumBlue size = 4> <b>Закон распределения вероятностей дискретной двумерной случайной величины</b> </font>

Зная закон распределения двумерной дискретной случайной величины, можно найти законы распределения каждой из составляющих. Действительно, например, события $(X=x_1;Y=y_1)$, $(X=x_1;Y=y_2)$, $\ldots$, $(X=x_1;Y=y_m)$ несовместны, поэтому вероятность $P(x_1)$ того, что $X$ примет значение $x_1$, по теореме сложения такова:\
$P(x_1) = p(x_1,y_1) + p(x_1,y_2) + \ldots + p(x_1,y_m)$.

Таким образом, для того чтобы найти вероятность $P(X=x_i)$, надо просуммировать вероятности "столбца $x_i$". Аналогично сложив вероятности "строки $y_j$", получим вероятность $P(Y=y_j)$.

### <font color = MediumBlue size = 4> <b>Функция распределения двумерной случайной величины</b> </font>

Рассмотрим двумерную случайную величину $(X,Y)$ (дискретную или непрерывную). Пусть $x,y$ — пара действительных чисел. Вероятность события, состоящего в том, что $X$ примет значение, меньшее $x$, и при этом $Y$ примет значение, меньшее $y$, обозначим через $F(x,y)$. Если $x$ и $y$ будут изменяться, то, вообще говоря, будет изменяться и $F(x,y)$, т.е. $F(x,y)$ есть 
функция от $x$ и $y$.

**Функцией распределения двумерной случайной величины** $(X,Y)$ называют функцию $F(x,y)$, определяющую для каждой пары чисел $x,y$ вероятность того, что $X$ примет значение, меньшее $x$, и при этом $Y$ примет значение, меньшее $y$:

$$F(x,y) = P(X<x,T<y)$$

Геометрически это равенство можно истолковать так: $F(x,y)$ есть вероятность того, что случайная точка $(X,Y)$ попадет в  
бесконечный квадрант с вершиной $(x,y)$, расположенный левее и ниже этой вершины.

#### <font color = SteelBlue size = 3> <b>Свойства функции распределения двумерной случайной величины</b> </font>

##### Свойство 1

Значения функции распределения удовлетворяют двойному неравенству: $0 \leq F(x,y) \leq 1$

Свойство вытекает из определения функции распределения как вероятности: вероятность — всегда неотрицательное число, не превышающее единицу.

##### Свойство 2

$F(x,y)$ есть неубываюьщя функция по каждому аргументу, т.е. \
$F(x_2,y) \geq F(x_1,y)$, если $x_2 > x_1$; \
$F(x,y_2) \geq F(x,y_1)$, если $y_2 > y_1$.

##### Свойство 3

Имеют место предельные соотношения:
1) $F(-\infty,y)=0$ \
2) $F(x,-\infty)=0$ \
3) $F(-\infty,-\infty)=0$ \
4) $F(\infty,\infty)=1$

##### Свойство 4

При $y=\infty$ функция распределения системы становится функцией распределения составляющей $X$: $F(x,\infty)=F_1(x)$.

При $x=\infty$ функция распределения системы становится функцией распределения составляющей $Y$: $F(\infty,y)=F_2(y)$.

___
### <font color = MediumBlue size = 4> <b>Условие независимости случайных величин</b> </font>

Для того чтобы случайные величины $X$ и $Y$ были независимыми, необходимо и достаточно, чтобы функция распределения системы $(X,Y)$ была равна произведению функций распределения составляющих:\
$F(x,y)=F_1(x)F_2(y)$

Для того чтобы непрерывные случайные величины $X$ и $Y$ были независимыми, необходимо и достаточно, чтобы плотность совместного распределения системы $(X,Y)$ была равна произведению плотностей распределения составляющих: \
$f(x,y)=f(x)f(y)$

### <font color = MediumBlue size = 4> <b>Условные законы распределения составляющих системы дискретных случайных величин</b> </font>

**Условным распределением составляющей** $X$ при $Y=y_j$ называют совокупность условных вероятностей $p(x_1|y_j),p(x_2|y_j),\ldots,p(x_n|y_j)$, вычисленных в предположении, что событие $Y=y_j$ уже наступило.  
Аналогично определяется условное распределение составляющей $Y$.

Зная закон распределения двумерной дискретной случайной величины, можно, пользуясь формулой $P(B|A) = \cfrac{P(AB)}{P(A)}$, вычислить условные законы распределения составляющих. 

В общем случае условные законы распределения составляющей X определяются соотношением: $P(x_i|y_j) = \cfrac{P(x_i,y_j)}{P(y_j)}$.

Аналогично находят условные законы распределения составляющей $Y$: $P(y_j|x_i) = \cfrac{P(x_i,y_j)}{P(x_i)}$.

> Сумма вероятностей условного распределения равна единице.

### <font color = MediumBlue size = 4> <b>Числовые характеристики системы двух случайных величин</b> </font>

#### <font color = SteelBlue size = 3> <b>Ковариация / Корреляционный момент</b> </font>

**Ковариацией** или **корреляционным моментом** $\mu_{xy}$ случайных величин $X$ и $Y$ называют математическое ожидание произведения отклонений этих величин:

$$\mu_{xy} = M(\,(X-M(X)) \cdot (Y-M(Y))\,)$$
<br><br>


Для вычисления корреляционного момента **дискретных** величин используют формулу:\
$\displaystyle \mu_{xy} = \sum_{i=1}^{n} \sum_{j=1}^{m} [x_i-M(X)] [y_j-M(Y)] p(x_i,y_j) = M(XY) - M(X)M(Y)$, где \
$\displaystyle M(XY) = \sum_{i=1}^{n} \sum_{j=1}^{m} x_i y_j p_{ij}$, \
$\displaystyle M(X) = \sum_{i=1}^{n}{x_i p_i}$, \
$\displaystyle M(Y) = \sum_{j=1}^{m}{y_j p_j}$
<br><br>


Для вычисления корреляционного момента **непрерывных** величин используют формулу:\
$\displaystyle \mu_{xy} = \int\limits_{-\infty}^{\infty} \int\limits_{-\infty}^{\infty} [x-M(X)] [y-M(Y)] f(x,y) dx dy$
<br><br>


**Ковариация – это мера линейной зависимости случайных величин.**

Ненулевое значение ковариации $\mu_{xy} \not= 0$ говорит нам о зависимости случайных величин $X$ и $Y$, причём, чем оно больше *по модулю*, тем эта зависимость ближе к функциональной линейной зависимости $y=kx+b$.

Если установлено, что корреляционный момент $\mu_{xy}=0$, то случайные величины могут оказаться как **независимыми**, так и **зависимыми** (т.к. зависимость может носить не только линейный характер).\
Но если известно, что $X$ и $Y$ **независимы**, то корреляционный момент $\mu_{xy}=0$.
<br><br>


Коэффициент ковариации принимает значения, не превосходящие *по модулю* $\mu_{xy} \leq \sqrt{D(X) \cdot D(Y)}.$

#### <font color = SteelBlue size = 3> <b>Коэффициент корреляции Пирсона</b> </font>

**Коэффициентом корреляции** $r_{xy}$ случайных величин $X$ и $Y$ называют отношение корреляционного момента к произведению  
средних квадратических отклонений этих величин: \
$r_{xy} = \cfrac{\mu_{xy}}{\sigma_x \sigma_y}$

> Коэффициент корреляции независимых случайных величин равен нулю (т.к. $\mu_{xy}=0$).

<br>

**Коэффициент корреляции безразмерен и принимает значения из промежутка: $-1\leq r_{xy} \leq 1$.**

Чем ближе $r_{xy}$ *по модулю* к $1$, тем теснее линейная взаимосвязь между величинами $X$ и $Y$.

Чем ближе $r_{xy}$ *по модулю* к $0$ – тем такая зависимость выражена меньше.

Взаимосвязь считается существенной, начиная примерно $|r_{xy}|=0.7$.

Крайним значениям $r_{xy}=-1$ и $r_{xy}=1$ соответствует строгая функциональная зависимость $y=kx+b$.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
Y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])

m = np.mean(X*Y) - np.mean(X)*np.mean(Y)
SDx = np.std(X)
SDy = np.std(Y)
r = m / (SDx * SDy)

plt.scatter(X, Y)
plt.title(f'r={r:.3f}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

np.corrcoef(X, Y)
```

#### <font color = SteelBlue size = 3> <b>Коэффициент корреляции Спирмена</b> </font>

Коэффициент корреляции Спирмена называют ранговым коэффициентом корреляции. Он также показывает тесноту линейной связи, но в отличии от коэффициента корреляции Пирсона не требует нормальности распределений случайных величин и применяется для количественных и  порядковых данных.

$r_s = 1 - \cfrac{6 \sum{d_i^2}}{N(N^2-1)}$

```python
stats.spearmanr(X, Y)
```

```python
import pandas as pd

n = X.shape[0]
df = pd.DataFrame({'X': X, 'Y': Y})
df['Xr'] = df['X'].rank()
df['Yr'] = df['Y'].rank()
df['d2'] = (df['Xr'] - df['Yr'])**2
df
```

```python
1 - 6 * df['d2'].sum() / (n*(n**2-1))
```

### <font color = MediumBlue size = 4> <b>Линейная регрессия</b> </font>

Рассмотрим двумерную случайную величину $(X,Y)$, где $X$ и $Y$ — зависимые случайные величины. Представим одну из величин как 
функцию другой. Ограничимся приближенным представлением величины $Y$ в виде линейной функции величины $X$: \
$Y \approx g(X) = \alpha + \beta X$, где \
$\alpha$ и $\beta$ — параметры, подлежащие определению.

#### <font color = SteelBlue size = 3> <b>Метод наименьших квадратов</b> </font>

Функцию $g(x) = \alpha + \beta X$ называют "наилучшим приближением" $Y$ в смысле метода наименьших квадратов, если математическое ожидание $M[Y-g(X)]^2$ принимает наименьшее возможное значение; функцию $g(x)$ называют **среднеквадратической регрессией** $Y$ на $X$.

Подберем параметры $\alpha$ и $\beta$ так, чтобы сумма квадратов отклонений была минимальной (в этом состоит сущность метода наименьших квадратов). Так как каждое отклонение зависит от отыскиваемых параметров, то и сумма квадратов отклонений есть функция $F$ этих параметров:\
$\displaystyle F(\alpha,\beta) = \sum_{i=1}^{n}{(Y_i-y_i)^2} = \sum_{i=1}^{n}{(\alpha + \beta x_i -y_i)^2}$

Для отыскания минимума приравняем нулю соответствующие частные производные: \
$\displaystyle \cfrac{\partial{F}}{\partial{b}} = 2\sum_{i=1}^{n}{(\alpha+\beta x_i - y_i)x_i} = 0$ \
$\displaystyle \cfrac{\partial{F}}{\partial{a}} = 2\sum_{i=1}^{n}{(\alpha+\beta x_i - y_i)} = 0$


Линейная средняя квадратическая регрессия $Y$ на $X$ имеет вид
$$g(X) = m_y + r \cfrac{\sigma_y}{\sigma_x} (X - m_x)$$
где $m_x = M(X)$, $m_y = M(Y)$, $\sigma_x = \sqrt{D(X)}$, $\sigma_y = \sqrt{D(Y)}$, 
$r = \cfrac{\mu_{xy}}{\sigma_x\sigma_y}$ – коэффициент корреляции величин $X$ и $Y$.

***
$\alpha = m_y - r \cfrac{\sigma_y}{\sigma_x} m_x$

$\beta = r \cfrac{\sigma_y}{\sigma_x}$ – **коэффициент регрессии** $Y$ на $X$

***
$y - m_y = r \cfrac{\sigma_y}{\sigma_x} (X - m_x)$ – прямая среднеквадратической регрессии $Y$ на $X$, где \
$r \cfrac{\sigma_y}{\sigma_x}$ – коэффициент регрессии $Y$ на $X$, \
$\sigma_y^2(1-r^2)$ – **остаточная дисперсия** случайной величины $Y$ относительно случайной величины $X$ \
(характеризует величину ошибки, которую допускают при замене $Y$ линейной функцией $g(X) = \alpha + \beta X$).

***
$x - m_x = r \cfrac{\sigma_x}{\sigma_y} (Y - m_y)$ – прямая среднеквадратической регрессии $X$ на $Y$, \
$r \cfrac{\sigma_x}{\sigma_y}$ – коэффициент регрессии $X$ на $Y$, \
$\sigma_x^2(1-r^2)$ – остаточная дисперсия случайной величины $X$ относительно $Y$.

***
Если $r = \pm1$, то $X$ и $Y$ связаны линейной функциональной зависимостью и обе прямые регрессии, совпадают.

Обе прямые регрессии проходят через точку $(m_x;m_y)$, которую называют **центром совместного распределения** величин $X$ и $Y$.
